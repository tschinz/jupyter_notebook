## WordCloud from HTML

```python
# get the data once
!wget -O - 'http://www.sfgate.com' > ../sample_files/data/sfgate.html
```

```python
%matplotlib inline 

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from bs4 import BeautifulSoup
import nltk

%config InlineBackend.figure_format = 'retina'

html = open("../sample_files/data/sfgate.html",'r').read()
#text = nltk.clean_html(html)
soup = BeautifulSoup(html, "lxml")
text = soup.get_text()
cleaned = nltk.word_tokenize(text.lower())
wordlist = [x for x in cleaned if (len(x)>=2 and x.isalpha())]
wordcloud = WordCloud(width=800, height=400, stopwords=STOPWORDS, background_color='white').generate(" ".join(wordlist))
plt.figure(figsize=(30,20))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

## Wordcloud from Text

```python
%matplotlib inline 

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import nltk

#text = open("../sample_files/data/huck_finn.txt",'r').read()
text = open("../sample_files/data/william_shakespeare_complete.txt",'r').read()
cleaned = nltk.word_tokenize(text.lower())
wordlist = [x for x in cleaned if (len(x)>=2 and x.isalpha())]
wordcloud = WordCloud(width=800, height=400, stopwords=STOPWORDS, background_color='black').generate(" ".join(wordlist))
plt.figure(figsize=(30,20))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
