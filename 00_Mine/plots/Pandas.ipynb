```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.title("Title")
plt.show() 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 3*np.pi, 500)
plt.plot(x, np.sin(x**2))
plt.title('A simple chirp');

plt.show()
```
