# Bin => Hex => Dec

```python
binary = '10011010' # without 0b
nbr    = 10

dec = int(binary, 2)
print("   i | Hexadecimal |                           Binary |    Decimal")
print("-----|-------------|----------------------------------|-----------")
for i in range(nbr):
  print("{0:4} | {1:11x} | {1:032b} | {1:10d}".format(i,dec+i))
```

# Hex => Bin => Decimal

```python
hexa = 'a' # without 0x
nbr    = 10

dec = int(hexa, 16)
print("   i | Hexadecimal |                           Binary |    Decimal")
print("-----|-------------|----------------------------------|-----------")
for i in range(nbr):
  print("{0:4} | {1:11x} | {1:032b} | {1:10d}".format(i,dec+i))
```
