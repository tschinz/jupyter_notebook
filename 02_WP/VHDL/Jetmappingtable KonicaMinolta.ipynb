## Calc Nozzle Pixel Address

```python
def calc_pixel_address(flip_y=False, flip_x=False):
  print("+----------------------------------------+")
  if (flip_y and flip_x):
    print("|            Flip X and Flip Y           |")
  elif flip_y:
    print("|                 Flip Y                 |")
  elif flip_x:
    print("|                 Flip X                 |")
  else:
    print("|                 no Flip                |")
  print("+------------------------+---------------+")
  print("| Nozzle |    Channel L  |    Channel R  |")
  print("|        | Pos X | Pox Y | Pos X | Pos Y |")
  print("+--------+-------+-------+-------+-------+")

  for nozzle in range(512): # 0 - 511
    
    if flip_y == False:
      nozzle_address_ch0 = 511 - nozzle
      nozzle_address_ch1 = nozzle
      pos_y_ch0 = nozzle_address_ch0 * 2 + 1
      pos_y_ch1 = nozzle_address_ch1 * 2 + 0
    else:
      nozzle_address_ch0 = nozzle
      nozzle_address_ch1 = 511 - nozzle
      pos_y_ch0 = nozzle_address_ch0 * 2 + 0
      pos_y_ch1 = nozzle_address_ch1 * 2 + 1
    
    if flip_x == False:
      if (nozzle%2 == 0):
        pos_x_ch0 = 0 + 28 + 12
        pos_x_ch1 = 0
      else:
        pos_x_ch0 = 0 + 28
        pos_x_ch1 = 0 + 12
    else:
      if (nozzle%2 == 0):
        pos_x_ch0 = 0
        pos_x_ch1 = 0 + 28 + 12
      else:
        pos_x_ch0 = 0 + 12
        pos_x_ch1 = 0 + 28
      
    print("| {0:6} | {1:5} | {2:5} | {3:5} | {4:5} |".format(nozzle, pos_x_ch0, pos_y_ch0, pos_x_ch1, pos_y_ch1))

calc_pixel_address(flip_y = False, flip_x = False)
calc_pixel_address(flip_y = False, flip_x = True)
calc_pixel_address(flip_y = True, flip_x = False)
calc_pixel_address(flip_y = True, flip_x = True)
```

## KM1800i

### Calc Pixel Row Position

```python
def calc_pixel_row_position(mm_l_r, mm_a_b_c): # distance between LC & RA - distance between A & B or B & C
  resolution     = 600 # dpi
  inch2cm        = 2.54 # cm/inch
  pixel_pitch    = (inch2cm/float(resolution))*10 # in mm
  channels_mm = [[0,0,0],[0,0,0]] # LC,LB,LA & RA,RB,RC
  
  channels_mm[0][0] = (0)/pixel_pitch # LC
  channels_mm[0][1] = (channels_mm[0][0] + mm_a_b_c) # LB
  channels_mm[0][2] = (channels_mm[0][1] + mm_a_b_c) # LA
  channels_mm[1][0] = (channels_mm[0][2] + mm_l_r)   # RA
  channels_mm[1][1] = (channels_mm[1][0] + mm_a_b_c) # RB
  channels_mm[1][2] = (channels_mm[1][1] + mm_a_b_c) # RC
  
  channels_px = []
  for channel in range(len(channels_mm)):
    channels_px.append([])
    for row in range(len(channels_mm[channel])):
      channels_px[channel].append(channels_mm[channel][row]/pixel_pitch)
  
  print("Pixel Row position")
  print("Channel LC = {:>6.5} mm => {:>6.5} px => {:>3} px @ {}dpi".format(channels_mm[0][0], channels_px[0][0], int(round(channels_px[0][0])), resolution))
  print("Channel LB = {:>6.5} mm => {:>6.5} px => {:>3} px @ {}dpi".format(channels_mm[0][1], channels_px[0][1], int(round(channels_px[0][1])), resolution))
  print("Channel LA = {:>6.5} mm => {:>6.5} px => {:>3} px @ {}dpi".format(channels_mm[0][2], channels_px[0][2], int(round(channels_px[0][2])), resolution))
  print("Channel RA = {:>6.5} mm => {:>6.5} px => {:>3} px @ {}dpi".format(channels_mm[1][0], channels_px[1][0], int(round(channels_px[1][0])), resolution))
  print("Channel RB = {:>6.5} mm => {:>6.5} px => {:>3} px @ {}dpi".format(channels_mm[1][1], channels_px[1][1], int(round(channels_px[1][1])), resolution))
  print("Channel RC = {:>6.5} mm => {:>6.5} px => {:>3} px @ {}dpi".format(channels_mm[1][2], channels_px[1][2], int(round(channels_px[1][2])), resolution))
  
calc_pixel_row_position(mm_l_r=1.016, mm_a_b_c=0.931)
```

## Calc Nozzle Pixel Address

```python
def calc_pixel_address(flip_y=False, flip_x=False):
  print("+----------------------------------------------+")
  if (flip_y and flip_x):
    print("|               Flip X and Flip Y              |")
  elif flip_y:
    print("|                    Flip Y                    |")
  elif flip_x:
    print("|                    Flip X                    |")
  else:
    print("|                    no Flip                   |")
  print("+------------------------------+---------------+")
  print("| Nozzle | Row |    Channel L  |    Channel R  |")
  print("|        |     | Pos X | Pox Y | Pos X | Pos Y |")
  print("+--------+-----+-------+-------+-------+-------+")

  channels = 2   # L & R
  rows     = 3   # A & B & C
  nozzles  = 296 # 0 - 295
  row_pos  = [0, 22, 44, 68, 90, 112] # row x-pos in px
  
  for nozzle in range(rows*nozzles): # 0 - 887
    
    if flip_y == False:
      nozzle_address_ch_l = 887 - nozzle
      nozzle_address_ch_r = nozzle
      pos_y_ch_l = nozzle_address_ch_l * channels + 1
      pos_y_ch_r = nozzle_address_ch_r * channels + 0
    else:
      nozzle_address_ch_l = nozzle
      nozzle_address_ch_r = 887 - nozzle
      pos_y_ch_l = nozzle_address_ch_l * channels + 0
      pos_y_ch_r = nozzle_address_ch_r * channels + 1
    
    if flip_x == False:
      if (nozzle%rows == 0):
        pos_x_ch_l = row_pos[5] # 112
        pos_x_ch_r = row_pos[0] # 0
        channel = "C"
      elif (nozzle%rows == 1):
        pos_x_ch_l = row_pos[4] # 90
        pos_x_ch_r = row_pos[1] # 22
        channel = "B"
      else:
        pos_x_ch_l = row_pos[3] # 68
        pos_x_ch_r = row_pos[2] # 44
        channel = "A"
    else:
      if (nozzle%rows == 0):
        pos_x_ch_l = row_pos[0] # 0
        pos_x_ch_r = row_pos[5] # 112
        channel = "C"
      elif (nozzle%rows == 1):
        pos_x_ch_l = row_pos[1] # 22
        pos_x_ch_r = row_pos[4] # 90
        channel = "B"
      else:
        pos_x_ch_l = row_pos[2] # 44
        pos_x_ch_r = row_pos[3] # 68
        channel = "A"
      
    print("| {0:6} | {1:>3} |{2:6} | {3:5} | {4:5} | {5:5} |".format(nozzle, channel, pos_x_ch_l, pos_y_ch_l, pos_x_ch_r, pos_y_ch_r))
  
calc_pixel_address(flip_y = False, flip_x = False)
calc_pixel_address(flip_y = False, flip_x = True)
calc_pixel_address(flip_y = True, flip_x = False)
calc_pixel_address(flip_y = True, flip_x = True)
```
