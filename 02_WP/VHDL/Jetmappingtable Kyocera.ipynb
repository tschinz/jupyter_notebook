```python
def calc_nozzle_coord(unit=1, line_nbr = 1, nozzle_nbr = 1):
  y_u1 = {1 : 810,2 : 890,3 : 900,4 : 980,5 : 990,6 : 1070,7 : 1080,8 : 1160,9 : 1170,10 : 1250,11 : 1260,12 : 1340,13 : 1350,14 : 1430,15 : 1440,16 : 1520}
  y_u2 = {1 : 710,2 : 630,3 : 620,4 : 540,5 : 530,6 : 450,7 : 440,8 : 360,9 : 350,10 : 270,11 : 260,12 : 180,13 : 170,14 : 90,15 : 80,16 : 0}
  coord = [0,0] # (X, Y)
  # Get Y Value
  if unit == 1:
    coord[1] = y_u1[line_nbr]
  elif unit == 2:
    coord[1] = y_u2[line_nbr]
  
  # Calc X Value
  if unit == 1:
    if line_nbr == 1:
      coord[0] = 32*(nozzle_nbr-1)+9
    elif line_nbr == 2:
      coord[0] = 32*(nozzle_nbr-1)+24
    elif line_nbr == 3:
      coord[0] = 32*(nozzle_nbr-1)+7
    elif line_nbr == 4:
      coord[0] = 32*(nozzle_nbr-1)+28
    elif line_nbr == 5:
      coord[0] = 32*(nozzle_nbr-1)+11
    elif line_nbr == 6:
      coord[0] = 32*(nozzle_nbr-1)+22
    elif line_nbr == 7:
      coord[0] = 32*(nozzle_nbr-1)+5
    elif line_nbr == 8:
      coord[0] = 32*(nozzle_nbr-1)+30
    elif line_nbr == 9:
      coord[0] = 32*(nozzle_nbr-1)+13
    elif line_nbr == 10:
      coord[0] = 32*(nozzle_nbr-1)+20
    elif line_nbr == 11:
      coord[0] = 32*(nozzle_nbr-1)+3
    elif line_nbr == 12:
      coord[0] = 32*(nozzle_nbr-1)+32
    elif line_nbr == 13:
      coord[0] = 32*(nozzle_nbr-1)+15
    elif line_nbr == 14:
      coord[0] = 32*(nozzle_nbr-1)+18
    elif line_nbr == 15:
      coord[0] = 32*(nozzle_nbr-1)+1
    elif line_nbr == 16:
      coord[0] = 32*(nozzle_nbr-1)+26
  elif unit == 2:
    if line_nbr == 1:
      coord[0] = 32*(nozzle_nbr-1)+16
    elif line_nbr == 2:
      coord[0] = 32*(nozzle_nbr-1)+19
    elif line_nbr == 3:
      coord[0] = 32*(nozzle_nbr-1)+2
    elif line_nbr == 4:
      coord[0] = 32*(nozzle_nbr-1)+27
    elif line_nbr == 5:
      coord[0] = 32*(nozzle_nbr-1)+10
    elif line_nbr == 6:
      coord[0] = 32*(nozzle_nbr-1)+21
    elif line_nbr == 7:
      coord[0] = 32*(nozzle_nbr-1)+4
    elif line_nbr == 8:
      coord[0] = 32*(nozzle_nbr-1)+31
    elif line_nbr == 9:
      coord[0] = 32*(nozzle_nbr-1)+14
    elif line_nbr == 10:
      coord[0] = 32*(nozzle_nbr-1)+23
    elif line_nbr == 11:
      coord[0] = 32*(nozzle_nbr-1)+6
    elif line_nbr == 12:
      coord[0] = 32*(nozzle_nbr-1)+29
    elif line_nbr == 13:
      coord[0] = 32*(nozzle_nbr-1)+12
    elif line_nbr == 14:
      coord[0] = 32*(nozzle_nbr-1)+25
    elif line_nbr == 15:
      coord[0] = 32*(nozzle_nbr-1)+8
    elif line_nbr == 16:
      coord[0] = 32*(nozzle_nbr-1)+17
  
  # Print Coordinates
  print(" {0:4} | {1:4} | {2:6} | {3:4} | {4:4} |".format(unit, line_nbr, nozzle_nbr, coord[0], coord[1]))
      
def calc_all_nozzle_coord():
  print(" Unit | Line | Nozzle |    X |    Y |")
  print("------+------+--------+------+------+")
  for unit in range(1,3):         # 1 - 2
    for line in range(1,17):      # 1 - 16
      for nozzle in range(1,167): # 1 - 166
        calc_nozzle_coord(unit, line, nozzle)

calc_all_nozzle_coord()
```
