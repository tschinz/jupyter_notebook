# Xilinx Spartan3E DCM Calculation
Allow to calculate Multiplier and Divisor values for a given input and output clock

## Imports

## Configuration

```python
input_clk = 66e6
output_clk = 25.175e6
multiplier_range = range(1,32)
divisor_range = range(1,32)
```

## Calculation

```python
# Calculate all possible frequencies
all_possible_clk = []
for m in multiplier_range:
  for d in divisor_range:
    all_possible_clk.append([input_clk*m/d, m, d])

# find the closest one
idx_closest = min(range(len(all_possible_clk)), key=lambda i: abs([all_clk[0] for all_clk in all_possible_clk][i]-output_clk))
```

## Print results

```python
# print the closest
print("Closest Frequency to {} from {} is:".format(EngNumber(output_clk, precision=3), EngNumber(input_clk, precision=3)))
print("  Output Clock : {}".format(EngNumber(all_possible_clk[idx_closest][0], precision=3)))
print("  Multiplier   : {}".format(all_possible_clk[idx_closest][1]))
print("  Divisor      : {}".format(all_possible_clk[idx_closest][2]))
```

```python
# print all
for clk in all_possible_clk:
  print("{} * {} / {} = {}".format(EngNumber(input_clk, precision=3), clk[1], clk[2], EngNumber(clk[0], precision=3)))
```
